```python
import pandas as pd
import datetime
```

```python
df=pd.read_csv('test-data.csv')
```

```python
df.head()
```

```python
df['order_day']=pd.to_datetime(df['created'])
```

```python
df['order_week']=df['order_day'].dt.week
```

```python
#уникальных пользователей
df.user_id.nunique()
```

```python
df_users=pd.DataFrame(df.groupby('user_id').agg('order_week').min()).rename(columns={'order_week':'first_order_week'}) \
                    .reset_index()
```

```python
df=df.merge(df_users, left_on='user_id', right_on='user_id', how='outer')
```

```python
df['week_number_after_first_week']=df['order_week']-df['first_order_week']
```

```python
df_1_week=df.query('week_number_after_first_week==1')
df_2_week=df.query('week_number_after_first_week==2')
df_3_week=df.query('week_number_after_first_week==3')
df_4_week=df.query('week_number_after_first_week==4')
df_5_week=df.query('week_number_after_first_week==5')
df_6_week=df.query('week_number_after_first_week==6')
df_7_week=df.query('week_number_after_first_week==7')
df_8_week=df.query('week_number_after_first_week==8')
```

```python
res_payment=pd.DataFrame(df_1_week.groupby('first_order_week').agg('payment').sum()) \
            .reset_index() \
            .rename(columns={'payment':'1_week'})
```

```python
res_payment=res_payment.merge(pd.DataFrame(df_2_week.groupby('first_order_week').agg('payment').sum()) \
            .reset_index() \
            .rename(columns={'payment':'2_week'}), 
                  left_on='first_order_week', right_on='first_order_week', how='outer')
res_payment=res_payment.merge(pd.DataFrame(df_3_week.groupby('first_order_week').agg('payment').sum()) \
            .reset_index() \
            .rename(columns={'payment':'3_week'}), 
                  left_on='first_order_week', right_on='first_order_week', how='outer')
res_payment=res_payment.merge(pd.DataFrame(df_4_week.groupby('first_order_week').agg('payment').sum()) \
            .reset_index() \
            .rename(columns={'payment':'4_week'}), 
                  left_on='first_order_week', right_on='first_order_week', how='outer')
res_payment=res_payment.merge(pd.DataFrame(df_5_week.groupby('first_order_week').agg('payment').sum()) \
            .reset_index() \
            .rename(columns={'payment':'5_week'}), 
                  left_on='first_order_week', right_on='first_order_week', how='outer')
res_payment=res_payment.merge(pd.DataFrame(df_6_week.groupby('first_order_week').agg('payment').sum()) \
            .reset_index() \
            .rename(columns={'payment':'6_week'}), 
                  left_on='first_order_week', right_on='first_order_week', how='outer')
res_payment=res_payment.merge(pd.DataFrame(df_7_week.groupby('first_order_week').agg('payment').sum()) \
            .reset_index() \
            .rename(columns={'payment':'7_week'}), 
                  left_on='first_order_week', right_on='first_order_week', how='outer')
res_payment=res_payment.merge(pd.DataFrame(df_8_week.groupby('first_order_week').agg('payment').sum()) \
            .reset_index() \
            .rename(columns={'payment':'8_week'}), 
                  left_on='first_order_week', right_on='first_order_week', how='outer')
```

```python
data=pd.read_csv('test-data.csv')
```

```python
data.head()
```

```python
data['order_date']=pd.to_datetime(data['created'])
```

```python
data['order_month']=data['order_date'].dt.month
```

```python
data_users=pd.DataFrame(data.groupby('user_id').agg('order_month').min()).rename(columns={'order_month':'first_order_month'}) \
                    .reset_index()
data_users.head()
```

```python
data=data.merge(data_users, left_on='user_id', right_on='user_id', how='outer')
```

```python
data['month_number_after_first_month']=data['order_month']-data['first_order_month']
```

```python
data_1_month=data.query('month_number_after_first_month==1')
data_2_month=data.query('month_number_after_first_month==2')
data_3_month=data.query('month_number_after_first_month==3')
data_4_month=data.query('month_number_after_first_month==4')
data_5_month=data.query('month_number_after_first_month==5')
data_6_month=data.query('month_number_after_first_month==6')
data_7_month=data.query('month_number_after_first_month==7')
data_8_month=data.query('month_number_after_first_month==8')
data_9_month=data.query('month_number_after_first_month==9')
```

```python
res_month=pd.DataFrame(data_1_month.groupby('first_order_month').agg('user_id').count()) \
            .reset_index() \
            .rename(columns={'user_id':'1_month'})
```

```python
res_month=res_month.merge(pd.DataFrame(data_2_month.groupby('first_order_month').agg('user_id').count()) \
            .reset_index() \
            .rename(columns={'user_id':'2_month'}), left_on='first_order_month', right_on='first_order_month', how='outer')
res_month=res_month.merge(pd.DataFrame(data_3_month.groupby('first_order_month').agg('user_id').count()) \
            .reset_index() \
            .rename(columns={'user_id':'3_month'}), left_on='first_order_month', right_on='first_order_month', how='outer')
res_month=res_month.merge(pd.DataFrame(data_4_month.groupby('first_order_month').agg('user_id').count()) \
            .reset_index() \
            .rename(columns={'user_id':'4_month'}), left_on='first_order_month', right_on='first_order_month', how='outer')
res_month=res_month.merge(pd.DataFrame(data_5_month.groupby('first_order_month').agg('user_id').count()) \
            .reset_index() \
            .rename(columns={'user_id':'5_month'}), left_on='first_order_month', right_on='first_order_month', how='outer')
res_month=res_month.merge(pd.DataFrame(data_6_month.groupby('first_order_month').agg('user_id').count()) \
            .reset_index() \
            .rename(columns={'user_id':'6_month'}), left_on='first_order_month', right_on='first_order_month', how='outer')
res_month=res_month.merge(pd.DataFrame(data_7_month.groupby('first_order_month').agg('user_id').count()) \
            .reset_index() \
            .rename(columns={'user_id':'7_month'}), left_on='first_order_month', right_on='first_order_month', how='outer')
res_month=res_month.merge(pd.DataFrame(data_8_month.groupby('first_order_month').agg('user_id').count()) \
            .reset_index() \
            .rename(columns={'user_id':'8_month'}), left_on='first_order_month', right_on='first_order_month', how='outer')
res_month=res_month.merge(pd.DataFrame(data_9_month.groupby('first_order_month').agg('user_id').count()) \
            .reset_index() \
            .rename(columns={'user_id':'9_month'}), left_on='first_order_month', right_on='first_order_month', how='outer')
```
